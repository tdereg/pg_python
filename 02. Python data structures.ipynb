# Variables

## Types
| Name | Type | Description |
| --- | -- | - |
| Integer | int | whole numbers such as 3, 300, 200 |
| Floating point | float | Numbers with a decimal point 2.3, 4.6, 10.0 |
| Strings | str | Ordered sequences of character" "hello", 'hello world' | 
| Lists | list | Ordered sequence of objects [10, "hello", 12.3] |
| Dictionaries | dict | {"mykey":"value", "name":"Tom"} |
| Tuples | tup | ordered immutable (cannot be changed) sequences of objects (10, "hello", 11.2) |
| Sets | set | Unordered collection of unique objects {"a", "b"} | 
| Booleans | bool | Logical value indicating True or False |

## Strings

  - sequence of characters
  - single quotes or double quotes ('hello' or "hello")
  - each character has specific postion in String 
  - Indexing (get specific character)
     - String: "Tom"
     - Character: "T", "o", "m"
     - Index: 0 1 2
     - Reverse: 0 -2 -1
  - slicing
    - [start:stop:step]
    - start: index of the begging of slice
    - end: index you will go up to
    - step: the size of jump to you take
  

```python
name = "Tom"
print (name[1])
print (name[-1])
```

```python
print ("hello \t world \nnew line")
```

```python
len("hello world")
```

```python
test_string = "Hello world"
print ("second character: "+ test_string[1])
print ("last character: "+ test_string[-1])
```

```python
test_string = "123456789"
print ("from second string till the end: "+ test_string[2:])
print ("from the beggingin till third string: "+ test_string[:3])
print ("from second to third: "+ test_string[1:3])
print ("from beggign to end, ever ysecond character: "+ test_string[::2])
print ("revert string (negative step): "+ test_string[::-1])
```
